```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import stemgraphic
import seaborn as sns

```

# Plota Diagrama de Ramo de Folha

```python
np.random.seed(42) #cria semente de variabilidade

# Gerando uma série de alturas com distribuição normal
altura_media = 170 #definindo media de altura
desvio_padrao = 13 #definindo desvio padrao
tamanho_amostra = 150 # definindo tamanho da amostra

alturas = np.round(np.random.normal(loc=altura_media, scale=desvio_padrao, size=tamanho_amostra,),2)
altura_serie = pd.Series(alturas)
fig, ax = stemgraphic.stem_graphic(altura_serie, aggregation=False, alpha = 0.15)

#definindo título
plt.title('Ramo de Folhas de Alturas (cm)')

#definindo texto do ramo (centena e dezena)
plt.text(-3, 17, 'Centena/Dezena', fontsize=11, color="black",
         bbox=dict(facecolor='#FDE7EF'))

#definindo texto da folha (unidade)
plt.text(5, 17, 'Unidade', fontsize=10, color="black",
         bbox=dict(facecolor='#DEECFE'))

plt.savefig('distribuicao_f.png')
```

# Plota Histograma de Alturas

```python
#determinando o estilo grade branca
sns.set(style="darkgrid")

#determinando tamanho da imagem
plt.figure(figsize=(10, 6))

#plotando histograma
sns.histplot(altura_serie, bins=20, kde=True, color='skyblue')

# adicionando título dos eixos
plt.title('Histograma de Alturas')
plt.xlabel('Altura (cm)')
plt.ylabel('Frequência')
```
